```python
def Deflectionwithq(q,L,E,I): #defining function to get deflection of beam with uniform load
    """L = length of cantilever
    I = moment of inertia
    E = modulus of elasticity;"""
    D1 = (q*L**4)/(8*E*I)
    return (D1)
```

```python
def Deflectionwith(L,E,I,F,a): #defining function to get deflection of beam with uniform load
    """L = length of cantilever
    I = moment of inertia
    E = modulus of elasticity;"""
    b = L - a
    D2=(F*a**3)/(8*E*I)*(1+3*a/2*b) 
    return (D2) # return the value
```
